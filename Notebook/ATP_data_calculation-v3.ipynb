```python
#Outliers

import numpy as np
from scipy.stats import zscore

#Z score

#Numeric columns to check
numeric_cols = ['Winner_Rank', 'Loser_Rank',
                'Winner_Aces', 'Loser_Aces',
                'Winner_DoubleFaults', 'Loser_DoubleFaults',
                'Match_Duration_Minutes']

#Function to calculate Z-score outliers for any dataset
def detect_outliers_zscore(df_subset, threshold=3):
    df_numeric = df_subset[numeric_cols].dropna()
    z_scores = np.abs(zscore(df_numeric))
    outliers = (z_scores > threshold).any(axis=1)
    
    outlier_count = outliers.sum()
    df_outliers = df_subset.loc[df_numeric.index[outliers]]
    
    return outlier_count, df_outliers

#Best of 3 matches
outlier_count_best3, df_best3_outliers = detect_outliers_zscore(df_best3)
print(f"Number of outliers in Best of 3 dataset: {outlier_count_best3}")

#Best of 5 matches
outlier_count_best5, df_best5_outliers = detect_outliers_zscore(df_best5)
print(f"Number of outliers in Best of 5 dataset: {outlier_count_best5}")

#View first few outliers
print("\nSample Outliers in Best of 3:")
print(df_best3_outliers.head())

print("\nSample Outliers in Best of 5:")
print(df_best5_outliers.head())

```

```python
#Outliers with %

#Count total records
total_best3 = len(df_best3)
total_best5 = len(df_best5)

#Best of 3 matches
outlier_count_best3, df_best3_outliers = detect_outliers_zscore(df_best3)
percent_outliers_best3 = (outlier_count_best3 / total_best3) * 100
print(f"Number of outliers in Best of 3 dataset: {outlier_count_best3} out of {total_best3} ({percent_outliers_best3:.2f}%)")

#Best of 5 matches
outlier_count_best5, df_best5_outliers = detect_outliers_zscore(df_best5)
percent_outliers_best5 = (outlier_count_best5 / total_best5) * 100
print(f"Number of outliers in Best of 5 dataset: {outlier_count_best5} out of {total_best5} ({percent_outliers_best5:.2f}%)")

#View first few outliers
print("\nSample Outliers in Best of 3:")
print(df_best3_outliers.head())

print("\nSample Outliers in Best of 5:")
print(df_best5_outliers.head())

```
